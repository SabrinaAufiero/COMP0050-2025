# Linear regression of Synchrony’s (SYF) and Wal-Mart’s (WMT) daily returns against the SPY daily returns, using daily returns for the years 2018 and 2019.

```python
import yfinance as yf
```

```python
# Fetch historical data for SPY, SYF, and WMT
spy_data = yf.Ticker("SPY").history(start=startDate, end=endDate)
syf_data = yf.Ticker("SYF").history(start=startDate, end=endDate)
wmt_data = yf.Ticker("WMT").history(start=startDate, end=endDate)
```

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pct_change.html

Fractional change between the current and a prior element:

(current-previous)/previous

NOTE: Despite the name of this method, it calculates fractional change (also known as per unit change or relative change) and not percentage change. If you need the percentage change, multiply these values by 100.

```python
# Calculate daily returns
spy_returns = spy_data['Close'].pct_change().dropna()
syf_returns = syf_data['Close'].pct_change().dropna()
wmt_returns = wmt_data['Close'].pct_change().dropna()
```

```python
# Align the data by date
returns_df = pd.DataFrame({
    'SPY': spy_returns,
    'SYF': syf_returns,
    'WMT': wmt_returns
}).dropna()
```

```python
display(returns_df)
```

```python
# Create side-by-side plots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot SYF vs SPY
axes[0].scatter(returns_df['SPY'], returns_df['SYF'], color='black', alpha=0.6)
axes[0].set_title("SYF vs SPY Returns")
axes[0].set_xlabel("SPY Daily Returns")
axes[0].set_ylabel("SYF Daily Returns")

# Plot WMT vs SPY
axes[1].scatter(returns_df['SPY'], returns_df['WMT'], color='black', alpha=0.6)
axes[1].set_title("WMT vs SPY Returns")
axes[1].set_xlabel("SPY Daily Returns")
axes[1].set_ylabel("WMT Daily Returns")

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

Let us go with the assumption that we can estimate the true beta of a stock to the market; i.e., the estimation error of the beta doesn’t really matter. Then a simple decomposition of market return + stock-specific return gives us a great deal of information.

```python
# Prepare independent (X) and dependent (y) variables for regression
X = returns_df[['SPY']].values  # SPY returns as the independent variable
y_syf = returns_df['SYF'].values  # SYF returns as the dependent variable
y_wmt = returns_df['WMT'].values  # WMT returns as the dependent variable
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model_syf = LinearRegression()
model_syf.fit(X, y_syf)
print("SYF Regression Coefficients:")
print(f"Intercept: {model_syf.intercept_}, Slope: {model_syf.coef_[0]}")
```

```python
# Perform linear regression for WMT
model_wmt = LinearRegression()
model_wmt.fit(X, y_wmt)
print("WMT Regression Coefficients:")
print(f"Intercept: {model_wmt.intercept_}, Slope: {model_wmt.coef_[0]}")
```

```python
# Generate fit lines using the regression models
spy_range = np.linspace(returns_df['SPY'].min(), returns_df['SPY'].max(), 100).reshape(-1, 1)
syf_fit_line = model_syf.predict(spy_range)  # Fitted values for SYF based on SPY range
wmt_fit_line = model_wmt.predict(spy_range)  # Fitted values for WMT based on SPY range

# Create side-by-side plots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot SYF vs SPY with regression line
axes[0].scatter(returns_df['SPY'], returns_df['SYF'], color='black', alpha=0.6, label="SYF Returns")
axes[0].plot(spy_range, syf_fit_line, color='red', label=f"Fit: y={model_syf.intercept_:.4f} + {model_syf.coef_[0]:.4f}x", linewidth=2)
axes[0].set_title("SYF vs SPY Returns")
axes[0].set_xlabel("SPY Daily Returns")
axes[0].set_ylabel("SYF Daily Returns")
axes[0].legend()

# Plot WMT vs SPY with regression line
axes[1].scatter(returns_df['SPY'], returns_df['WMT'], color='black', alpha=0.6, label="WMT Returns")
axes[1].plot(spy_range, wmt_fit_line, color='red', label=f"Fit: y={model_wmt.intercept_:.4f} + {model_wmt.coef_[0]:.4f}x", linewidth=2)
axes[1].set_title("WMT vs SPY Returns")
axes[1].set_xlabel("SPY Daily Returns")
axes[1].set_ylabel("WMT Daily Returns")
axes[1].legend()

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

For the overwhelming majority of stocks, the beta to the market is positive, but it can take values in a wide interval; several stocks exhibit betas higher than two.

# But now, the confidence interval ?

Hogg, McKean, and Craig, Introduction to Mathematical Statistics, 6th ed.

DEF: A $100 (1-\alpha)%$ confidence interval for a parameter $\theta$ is a random interval $[L_1, L_2]$ s.t. 

$P[L_1 \leq \theta \leq L_2] = 1 - \alpha$

regardelss of the value of $\theta$.

Confidence intervals provide a range of values within which the true value of a parameter (e.g., slope or intercept) is likely to lie with a specified level of confidence, such as 95%.

The interval is constructed in such a way that, if we were to repeat the sampling process many times, the true parameter would fall within this interval approximately 95% of the time (for a 95% confidence level).

For example: if a confidence interval for $\beta$ includes 0, it suggests that there may be no linear relationship between the independent and dependent variables.

```python
import statsmodels.api as sm
```

When using Scikit-Learn’s LinearRegression, the intercept is handled automatically -- so you don’t have to add a constant manually. However, with Statsmodels’ OLS, you must add a constant manually if you want an intercept in your model.

```python
# Independent and dependent variables
X = returns_df['SPY']
y_syf = returns_df['SYF']

# Add a constant term so we can estimate the intercept
X_sm = sm.add_constant(X)

# Fit the OLS regression model using Statsmodels
model = sm.OLS(y_syf, X_sm).fit()

# Extract coefficients and their 95% confidence intervals
alpha_est = model.params['const']
beta_est  = model.params['SPY']
```

- alpha_est is the estimated intercept.
- beta_est is the estimated slope coefficient, indicating how much SYF's returns change with a one-unit change in SPY's returns.

```python
ci = model.conf_int(alpha=0.05)
alpha_ci = ci.loc['const'].values  # [lower bound, upper bound]
beta_ci  = ci.loc['SPY'].values

print(alpha_est)
print(alpha_ci)
```

```python
# Annualize the intercept (alpha):
# Assuming 252 trading days per year and converting to percentage, annualized alpha (%) = daily alpha * 252 * 100.
alpha_est_ann = alpha_est * 252 * 100
alpha_ci_ann  = alpha_ci  * 252 * 100
```

```python
# Print the formatted table
print("Table 3.1 Parameter estimates for Synchrony’s returns regressed against SP500 returns, 2018. Alphas are expressed in percentage annualized returns.\n")

print(f"{'Parameter':<12}{'Estimate':>12}{'95% conf.interval':>25}")
print(f"{'alpha (%)':<12}{alpha_est_ann:>12.0f}{'(' + f'{alpha_ci_ann[0]:.0f}, {alpha_ci_ann[1]:.0f}' + ')':>25}")
print(f"{'beta':<12}{beta_est:>12.2f}{'(' + f'{beta_ci[0]:.2f}, {beta_ci[1]:.2f}' + ')':>25}")
```

```python
# Independent and dependent variables
X = returns_df['SPY']
y_wmt = returns_df['WMT']

# Add a constant term so we can estimate the intercept
X_sm = sm.add_constant(X)

# Fit the OLS regression model using Statsmodels
model = sm.OLS(y_wmt, X_sm).fit()

# Extract coefficients and their 95% confidence intervals
alpha_est = model.params['const']
beta_est  = model.params['SPY']
```

```python
ci = model.conf_int(alpha=0.05)
alpha_ci = ci.loc['const'].values  # [lower bound, upper bound]
beta_ci  = ci.loc['SPY'].values
```

```python
alpha_est_ann = alpha_est * 252 * 100
alpha_ci_ann  = alpha_ci  * 252 * 100
```

```python
print("Table 3.1 Parameter estimates for Walmart’s returns regressed against SP500 returns, 2018. Alphas are expressed in percentage annualized returns.\n")

print(f"{'Parameter':<12}{'Estimate':>12}{'95% conf.interval':>25}")
print(f"{'alpha (%)':<12}{alpha_est_ann:>12.0f}{'(' + f'{alpha_ci_ann[0]:.0f}, {alpha_ci_ann[1]:.0f}' + ')':>25}")
print(f"{'beta':<12}{beta_est:>12.2f}{'(' + f'{beta_ci[0]:.2f}, {beta_ci[1]:.2f}' + ')':>25}")
```

The error around the $\alpha$ estimate is much larger than the estimate itself, which is not the case for $\beta$.
